## Justification for the raster based approach in `yield_change_calculation.py`

### Context

In my main analysis code, I compare individual raster values against the country level boundaries. This is much slower than the approach used elsewhere in ALLFED's code, but my initial tests suggested that adopting this approach in my code wouldn't work (because the resolution of the raster in this analysis is so coarse, it would miss out 98 countries). Below I demonstrate this to be the case.

In addition, I also show that we need to map names from the country shapefile used elsewhere in ALLFED's codebase and that found in `rutgers_nw_production_raw.csv`, the reference file against which this code is validated and tested.

### Country shapefile

For consistency with other parts of ALLFED's code, I use country-level boundaries adopted from the shapefile used in https://github.com/allfed/LosingIndustryCropYields, `World_Countries__Generalized_`, here found in the directory `data/external/World_Countries__Generalized_`.

Below, I quickly inspect some of the properties of this shapefile.

```python
import geopandas as gpd

# File path for the shapefile
shapefile_path = 'data/external/World_Countries__Generalized_/World_Countries__Generalized_.shp'

# Load the shapefile with geopandas
gdf = gpd.read_file(shapefile_path)

# Print the first few rows of the GeoDataFrame to inspect the data
print("First few rows of the GeoDataFrame:")
print(gdf.head())

# Print the structure and data types of the GeoDataFrame
print("\nGeoDataFrame structure:")
print(gdf.info())

# Print the available columns
print("\nAvailable columns in the shapefile:")
print(gdf.columns)

# Print the coordinate reference system (CRS)
print("\nCoordinate Reference System (CRS):")
print(gdf.crs)

# Print the geometry type (Polygon/MultiPolygon)
print("\nGeometry type:")
print(gdf.geom_type.unique())

# Print the total number of countries or regions in the shapefile
print("\nTotal number of countries or regions:")
print(len(gdf))
```

Compared to the three letter `ISO 3 Country Code` values in `rutgers_nw_production_raw.csv`, we can also see that this shapefile has two letter ISO codes. I therefore use the ISO codes found in this shapefile throughout this codebase (though this can be revisited in later code).

### Intersecting raster points and country boundaries:

In [one piece of ALLFED's codebase](https://github.com/allfed/LosingIndustryCropYields/blob/3dc7554a6ce796e9a95724f66647d0242e91e21e/scripts/yield_decline_by_country.ipynb#L61) which aims to achieve a similar purpose, the join between a raster of crop yields and countries is handled similarly to the code below, i.e. converting the raster to a set of points, then merging based on which points fall within the boundaries of countries. However, based on this code example, we find that 98 countries (which are listed) end up without any raster points corresponding to them. Even though the point-based approach is fast compared to clipping the raster in Python, I opt for the latter in order to include the 98 countries that would otherwise go missing.

```python
import xarray as xr
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
import pandas as pd
import rioxarray

# File paths for one of the rasters and the countries shapefile
file_path = '../data/raw/Crop Yield/clm5_crop_2deg_cpl_nw_ur_150_07_0005-0019_yield_latlon_CLM5mask_6crops.nc'
shapefile_path = '../data/external/World_Countries__Generalized_/World_Countries__Generalized_.shp'

# Load the NetCDF dataset for the example file
ds = xr.open_dataset(file_path)

# Correct longitude from [0, 360] to [-180, 180] and sort
ds = ds.assign_coords(lon=(((ds['lon'] + 180) % 360) - 180)).sortby('lon')

# Load the shapefile with geopandas
gdf = gpd.read_file(shapefile_path)

# Ensure the CRS is consistent (EPSG:4326)
gdf = gdf.to_crs(epsg=4326)

# Extract crop names from the 'crops' coordinate attribute
crop_names = ds['crops'].attrs['long_name'].split(',')

# Select a specific crop (e.g., CornRain) and a specific year
crop = 'CornRain'
year = 0  # First year of data

# Find the index for the specific crop
crop_idx = crop_names.index(crop)

# Extract the yield data for the selected crop and year
yield_data = ds['yield'].isel(crops=crop_idx, time=year).fillna(0)

# Convert the yield data to a DataFrame
yield_df = yield_data.to_dataframe(name='yield').reset_index()

# # Remove NaN values (which indicate missing yield data)
# yield_df = yield_df.dropna()

# Convert the yield DataFrame to a GeoDataFrame with geometry (points from lat/lon)
yield_gdf = gpd.GeoDataFrame(
    yield_df, 
    geometry=gpd.points_from_xy(yield_df['lon'], yield_df['lat']),
    crs="EPSG:4326"
)

# Plot the points and the country boundaries to see how they intersect
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the country boundaries
gdf.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=0.5)

# Plot the yield points
yield_gdf.plot(ax=ax, color='red', markersize=1, alpha=0.6)

# Add title and labels
plt.title(f'CornRain Yield Points for Year {year + 1}')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.savefig('cornrain_yield_points_overlay.png')
plt.show()

# Find the countries that do not have any points within them
countries_with_points = gpd.sjoin(gdf, yield_gdf, how="inner", predicate='intersects')['COUNTRY'].unique()
missing_countries = set(gdf['COUNTRY'].unique()) - set(countries_with_points)

# Print the missing countries
print("Countries without any yield points:")
for country in missing_countries:
    print(country)

# Close the dataset after use
ds.close()
```

## Country renaming

Another addition to the configuration file is a map to rename countries as named in `World_Countries__Generalized_.shp` to their equivalents identified in `rutgers_nw_production_raw.csv`. This is necessary because as we can see in the code below, some countries are named differently between them.

```python
original = pd.read_csv('../data/raw/rutgers_nw_production_raw.csv')

original_countries = original['Country'].unique()
new_countries = gdf['COUNTRY'].unique()

missing1 = set(original_countries) - set(new_countries)
missing2 = set(new_countries) - set(original_countries)

print("Countries in the Rutgers study file not found in Shapefile:")
print(missing1)

print("Countries in the Shapefile not found in Rutgers output")
print(missing2)
```

Some of these are clearly the same, hence I map between them in the `config/config.yaml` file.
